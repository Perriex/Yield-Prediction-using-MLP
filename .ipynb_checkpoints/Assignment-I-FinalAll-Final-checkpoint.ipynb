## Neural Network Functions

```python
#Sigmoid Function
def sigmoid (x):
    return 1/(1 + np.exp(-x))

#derivative of Sigmoid Function
def derivatives_sigmoid(x):
    return x * (1 - x)
```

```python
def initialize_weights(dimensions):
    #initiliazing the random weights
    #weights for hidden layer
    weights_hidden = np.random.rand(dimensions[0], dimensions[1])
    #bias for hidden layer
    bias_hidden = np.random.rand(1, dimensions[1])
    #weights for output layer
    weights_output = np.random.rand(dimensions[1], dimensions[2])
    #bias for output layer
    bias_output = np.random.rand(1, dimensions[2])
    
    return (weights_hidden, bias_hidden, weights_output, bias_output)
```

```python
def forward_propagation(X, weights_hidden, bias_hidden, weights_output, bias_output):
    
    #Multiplying input with hidden layer weights and adding bias to get sum of hidden layer
    Sum_hidden_1 = np.dot(X, weights_hidden) + bias_hidden
    #finding the sigmoid of above output which is output of hidden layer
    hidden_1_output = sigmoid(Sum_hidden_1)
    #Multplying the output of hidden layer with output weights and adding bias to get sum of output layer 
    Sum_output_2 = np.dot(hidden_1_output, weights_output) + bias_output
    #finding the sigmoid of above output which is output of output layer
    y_output = sigmoid(Sum_output_2)
    
    return (y_output, hidden_1_output, weights_output)
```

```python
def backward_propagation(y, y_output, hidden_1_output, weights_output, lr, bias_output, weights_hidden, bias_hidden, special_case=False):    
    
    #computing the error
    error = y-y_output
    #finding the derivative of output from forward propagation
    d_sum_output = derivatives_sigmoid(y_output)
    #finding the derivative of output from hidden layer of forward propagation
    d_sum_hidden = derivatives_sigmoid(hidden_1_output)
    #finding the delta of output layer
    delta_output_layer = error * d_sum_output
    #finding the error in hidden layer
    error_hidden_layer = delta_output_layer.dot(weights_output.T)
    #finding the delta of hidden layer
    delta_hidden_layer = error_hidden_layer * d_sum_hidden
    
    #updating new weights
    weights_output = weights_output + lr * hidden_1_output.T.dot(delta_output_layer)
    bias_output = bias_output + lr * np.sum(delta_output_layer, axis=0,keepdims=True)
    weights_hidden = weights_hidden + lr * X.T.dot(delta_hidden_layer)
    #print(delta_hidden_layer.shape)
    bias_hidden = bias_hidden + lr * np.sum(delta_hidden_layer, axis=0,keepdims=True)
    
    return weights_hidden, bias_hidden, weights_output, bias_output, error
```

```python
def run_neuralnet(X, y, dimensions=[2,2,1], epochs=10000, lr=0.1):
    #initilize the weights based on dimensions
    weights_hidden, bias_hidden, weights_output, bias_output = initialize_weights(dimensions=dimensions)
    #create empty list to save the errors in each epoch
    error_list =[]
    print('Intial Weights:','\n')
    print('Weights_hidden :','\n', weights_hidden)
    print('Bias_hidden :','\n', bias_hidden)
    print('Weights_output :','\n', weights_output)
    print('Bias_output :','\n', bias_output)
    #for each epoch calculate forward and backward propagation and error
    for i in range(epochs):
        y_output, hidden_1_output, weights_output = forward_propagation(X, weights_hidden, bias_hidden, weights_output, bias_output)

        weights_hidden, bias_hidden, weights_output, bias_output, error = backward_propagation(y, y_output, hidden_1_output, weights_output, lr,bias_output, weights_hidden, bias_hidden, special_case=False)
        
        error_list.append(np.sum(np.square(error))/2)
    return weights_hidden, bias_hidden, weights_output, bias_output, error_list
```

```python
def test(X, weights_hidden, bias_hidden, weights_output, bias_output):
    #this is nothing but forward propagation with less outputs for testing
    Sum_hidden_1 = np.dot(X, weights_hidden)
    Sum_hidden_1 = Sum_hidden_1 + bias_hidden
    hidden_1_output = sigmoid(Sum_hidden_1)
    Sum_output_2 = np.dot(hidden_1_output, weights_output)
    Sum_output_2 = Sum_output_2 + bias_output
    y_output = sigmoid(Sum_output_2)
    
    return y_output
```
